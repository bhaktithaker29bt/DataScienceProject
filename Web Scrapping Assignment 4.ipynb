* Q1- Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views.

```python
# Activating the chrome browser
driver = webdriver.Chrome(r"C:\Users\bhakti.thaker\Desktop\Datatrained\FLIPROBBO\Selenium Web sCrapping\chromedriver_win32\chromedriver.exe")
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

time.sleep(2)
```

```python
#Extracting all tags for Ranks
rank_tags = driver.find_elements_by_xpath("//tbody/tr/td[1][@align = 'center']")
rank_tags
```

```python
# will extract the text from the tags.
#so will run the loop to iterate over the tags extracted above and extract the text inside them.
rank_titles=[]
for i in rank_tags:
    rank_titles.append(i.text)
rank_titles
```

```python
#Extracting all tags for Ranks
name_tags = driver.find_elements_by_xpath("//tbody/tr/td[2]/a")
name_tags
```

```python
# will extract the text from the tags.
#so will run the loop to iterate over the tags extracted above and extract the text inside them.
name_titles=[]
for i in name_tags:
    name_titles.append(i.text)
name_titles
```

```python
#Extracting all tags for Ranks
artist_tags = driver.find_elements_by_xpath("//tbody/tr/td[3]")
artist_tags
```

```python
# will extract the text from the tags.
#so will run the loop to iterate over the tags extracted above and extract the text inside them.
artist_titles=[]
for i in artist_tags:
    artist_titles.append(i.text)
artist_titles
```

```python
#Extracting all tags for Ranks
date_tags = driver.find_elements_by_xpath("//tbody/tr/td[5][@align = 'right']")
date_tags
```

```python
# will extract the text from the tags.
#so will run the loop to iterate over the tags extracted above and extract the text inside them.
date_titles=[]
for i in date_tags:
    date_titles.append(i.text)
date_titles
```

```python
#Extracting all tags for Ranks
views_tags = driver.find_elements_by_xpath("//tbody/tr/td[4][@align = 'center']")
views_tags
```

```python
# will extract the text from the tags.
#so will run the loop to iterate over the tags extracted above and extract the text inside them.
views_titles=[]
for i in views_tags:
    views_titles.append(i.text)
views_titles
```

```python
# will check number of data extracted
print(len(rank_titles),len(name_titles),len(artist_titles),len(date_titles),len(views_titles))
```
